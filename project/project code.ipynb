```python
dataPR = data[data["Pregnancy Risk"]!="No"]
print("Data base with molecules dangerous for pregnant women")
dataPR
```

```python
dataPRloreal = dataPR[dataPR["Cream"]=="L'Oréal"]
dataPRloreal
```

```python

from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display

# Assurez-vous que la colonne SMILES est correctement identifiée
smiles_data = dataPRloreal['SMILES']
print(smiles_data)# Remplacez 'SMILES' par le nom réel de votre colonne de SMILES



# Convertir les SMILES en molécules
molecules = [Chem.MolFromSmiles(smile) if pd.notna(smile) else None for smile in smiles_data]

# Filtrer les molécules None (au cas où il y a des SMILES non valides)
molecules = [mol for mol in molecules if mol is not None]

# Générer et afficher une grille d'images des molécules
if molecules:  # Vérifie que la liste n'est pas vide
    img = Draw.MolsToGridImage(molecules, molsPerRow=4, subImgSize=(200, 200))
    display(img)
else:
    print("Aucune molécule valide à afficher.")
```

```python
dataPRloreal = dataPR[dataPR["Cream"]=="L'Oréal"]
dataPRloreal
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display
import pandas as pd  # Assurez-vous que pandas est importé si ce n'est pas déjà fait

# Assurez-vous que les colonnes SMILES et Molecule sont correctement identifiées
smiles_data = dataPRloreal['SMILES']
names_data = dataPRloreal['Molecule']  # Remplacez 'Molecule' par le nom réel de votre colonne de noms

# Convertir les SMILES en molécules
molecules = [Chem.MolFromSmiles(smile) if pd.notna(smile) else None for smile in smiles_data]

# Extraire les noms associés uniquement aux molécules valides en utilisant iloc pour l'accès positionnel
valid_names = [names_data.iloc[i] for i, mol in enumerate(molecules) if mol is not None]

# Filtrer les molécules None (au cas où il y a des SMILES non valides)
molecules = [mol for mol in molecules if mol is not None]

# Générer et afficher une grille d'images des molécules avec les noms comme légendes
if molecules:  # Vérifie que la liste n'est pas vide
    img = Draw.MolsToGridImage(molecules, molsPerRow=4, subImgSize=(200, 200), legends=valid_names)
    display(img)
else:
    print("Aucune molécule valide à afficher.")
```
