```python
dataPR = data[data["Pregnancy Risk"]!="No"]
print("Data base with molecules dangerous for pregnant women")
dataPR
```

```python
dataPRloreal = dataPR[dataPR["Cream"]=="L'OrÃ©al"]
dataPRloreal
```

```python

from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display

# Assurez-vous que la colonne SMILES est correctement identifiÃ©e
smiles_data = dataPRloreal['SMILES']
print(smiles_data)# Remplacez 'SMILES' par le nom rÃ©el de votre colonne de SMILES



# Convertir les SMILES en molÃ©cules
molecules = [Chem.MolFromSmiles(smile) if pd.notna(smile) else None for smile in smiles_data]

# Filtrer les molÃ©cules None (au cas oÃ¹ il y a des SMILES non valides)
molecules = [mol for mol in molecules if mol is not None]

# GÃ©nÃ©rer et afficher une grille d'images des molÃ©cules
if molecules:  # VÃ©rifie que la liste n'est pas vide
    img = Draw.MolsToGridImage(molecules, molsPerRow=4, subImgSize=(200, 200))
    display(img)
else:
    print("Aucune molÃ©cule valide Ã  afficher.")
```

```python
dataPRloreal = dataPR[dataPR["Cream"]=="L'OrÃ©al"]
dataPRloreal
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
from IPython.display import display
import pandas as pd  # Assurez-vous que pandas est importÃ© si ce n'est pas dÃ©jÃ  fait

# Assurez-vous que les colonnes SMILES et Molecule sont correctement identifiÃ©es
smiles_data = dataPRloreal['SMILES']
names_data = dataPRloreal['Molecule']  # Remplacez 'Molecule' par le nom rÃ©el de votre colonne de noms

# Convertir les SMILES en molÃ©cules
molecules = [Chem.MolFromSmiles(smile) if pd.notna(smile) else None for smile in smiles_data]

# Extraire les noms associÃ©s uniquement aux molÃ©cules valides en utilisant iloc pour l'accÃ¨s positionnel
valid_names = [names_data.iloc[i] for i, mol in enumerate(molecules) if mol is not None]

# Filtrer les molÃ©cules None (au cas oÃ¹ il y a des SMILES non valides)
molecules = [mol for mol in molecules if mol is not None]

# GÃ©nÃ©rer et afficher une grille d'images des molÃ©cules avec les noms comme lÃ©gendes
if molecules:  # VÃ©rifie que la liste n'est pas vide
    img = Draw.MolsToGridImage(molecules, molsPerRow=4, subImgSize=(200, 200), legends=valid_names)
    display(img)
else:
    print("Aucune molÃ©cule valide Ã  afficher.")
```
