```python
dataPR = data[data["Pregnancy Risk"]!="No"]
print("Data base with molecules dangerous for pregnant women")
dataPR
```

```python
dataPRlaneige = dataPR[dataPR["Cream"]=="La Neige"]
dataPRlaneige 
```

```python
dataPR['Molecule'] = None

# Boucle Ã  travers les lignes de votre DataFrame et convertissez les SMILES en objets Mol RDKit
for index, row in dataPR.iterrows():
    smiles = row['SMILES']
    mol = Chem.MolFromSmiles(smiles)
    # Assurez-vous que la conversion a rÃ©ussi avant d'assigner le rÃ©sultat Ã  la colonne 'Molecule'
    if mol is not None:
        dataPR.at[index, 'Molecule'] = mol
```

```python
nb_lignes, nb_colonnes = dataPR.shape
for i in nb_lignes:
    dataPRtemp = dataPRlaneige.iloc[i]
    smiles= dataPRtemp['SMILES']
    mol = Chem.MolFromSmiles(smiles)
    mol
```
